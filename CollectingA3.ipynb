```python
import pandas as pd 
```

```python
raw = pd.read_csv("raw_bechdel.csv") #loading the dataset that I'll be working with 
raw
```

## Movies that were released after the year 2000 

As you can see in the table below there are 5391 movies that were released after 2000

```python
movies_after_two_thousand = raw[raw.year > 2000]
movies_after_two_thousand
```

## Movies with just 3 stars after the year 2000

As you can see in the table below there are 3367 movies that are rated 3 stars and were after the year 2001

```python
movies_rating = raw[(raw.year > 2000) & (raw.rating == 3)] 
movies_rating
```

## Find all the movies that have the name "Raya" in the title. 

As you can see in the table below there is 1 movie with the name "Raya" in it. 

```python
movies_raya = raw[raw.title.str.contains("Raya")]
movies_raya
```

## Show all the movies that were released between the 60s and 70s

As you can see in the table below there is 744 movies between the 60s and 70s

```python
movies_between = raw[(raw.year >= 1960) & (raw.year < 1980)] 
movies_between
```

## Show the decade with the best average of movie ratings 

As you can see below 2020 was the best decade.  

```python
start = 1880 #starting decade 
end = 2021 #end of decade 
ratings = {}
for i in range (2021//10):
    decade_movies = raw[(raw.year >= start) & (raw.year <= start + 9)] 
    sum_decade = decade_movies['rating'].sum()
    ratings[start] = sum_decade/len(decade_movies)
    start += 10
    if start > end:
        break
maximum_value = max(ratings, key=ratings.get) 
print(maximum_value)
```
